```
from PIL import Image
from PIL import ImageFilter
import os
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

inPath ='/content/pix2pixHD/datasets/w/test_A_original'
outPath ='/content/pix2pixHD/datasets/w/test_A'
for imagePath in os.listdir(inPath):
  inputPath = os.path.join(inPath, imagePath)
  # img = Image.open(inputPath)
  img=cv2.imread(inputPath)
  filtered  = cv2.GaussianBlur(img,(199,199),500)
  fullOutPath = os.path.join(outPath, imagePath)
  # cv2_imshow(filtered)
  cv2.imwrite(fullOutPath, filtered)
  print(fullOutPath)
```
